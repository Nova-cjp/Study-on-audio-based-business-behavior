```python
import re
import akshare as ak
from datetime import datetime, timedelta
prog=re.compile(r'[(](.*?)[)]')
with open(r"D:\Graduation_project\video\Release_Roadshow\show_date.txt","r")as f:
    stock_list=[]
    all_data=f.readlines()
for data in all_data:
    company_number=prog.findall(data)[0]
    date = datetime.strptime(data[:10],"%Y-%m-%d")
    date_str=datetime.strftime(date, "%Y%m%d")
    stock_zh_a_hist_df = ak.stock_zh_a_hist(symbol=company_number, period="daily", start_date=date_str, end_date=date_str, adjust="")
    while stock_zh_a_hist_df.empty is True:
        date=date+timedelta(days=1)
        date_str=datetime.strftime(date, "%Y%m%d")
        stock_zh_a_hist_df = ak.stock_zh_a_hist(symbol=company_number, period="daily", start_date=date_str, end_date=date_str, adjust="")
    first_end_price=stock_zh_a_hist_df["收盘"][0]
    date_after_30days=date+timedelta(days=30)
    date_after_30days_str=datetime.strftime(date_after_30days, "%Y%m%d")
    stock_zh_a_hist_df = ak.stock_zh_a_hist(symbol=company_number, period="daily", start_date=date_after_30days_str, end_date=date_after_30days_str, adjust="")
    while stock_zh_a_hist_df.empty is True:
        date_after_30days=date_after_30days+timedelta(days=1)
        date_after_30days_str=datetime.strftime(date_after_30days, "%Y%m%d")
        stock_zh_a_hist_df = ak.stock_zh_a_hist(symbol=company_number, period="daily", start_date=date_after_30days_str, end_date=date_after_30days_str, adjust="")    
    end_price=stock_zh_a_hist_df["收盘"][0]
    print((company_number,first_end_price,end_price))
    stock_list.append((company_number,first_end_price,end_price))
print(stock_list)
```

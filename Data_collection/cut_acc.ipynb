```python
from pydub import AudioSegment
from pydub.utils import make_chunks
import os
for root,dirs,files in os.walk(r"D:\Graduation_project\audio\Release_Roadshow"):
    for file in files:
        audio = AudioSegment.from_file(os.path.join(root,file), "aac")

        size = 40000  ##

        chunks = make_chunks(audio, size)  ##10s
        os.makedirs(os.path.join(root,'section'))
        for i, chunk in enumerate(chunks):
            ##ichunk
            chunk_name = "section{0}.wav".format(i)
            print(chunk_name)
            ##
            chunk.export(os.path.join(root,'section',chunk_name), format="wav", parameters=["-ac", "1", "-ar", "16000"])
#             import subprocess
#             subprocess.Popen(r'ffmpeg -y  -i '+os.path.join(root,'section',chunk_name)+'  -acodec pcm_s16le -f s16le -ac 1 -ar 16000 '+os.path.join(root,'section',chunk_name[:-3]+'pcm'), shell = False)
#             subprocess.Popen(r'del '+os.path.join(root,'section',chunk_name), shell = False)
print('finish')



# ##blues30s
# audio = AudioSegment.from_file("blues.00000.wav", "wav")

# size = 10000  ##

# chunks = make_chunks(audio, size)  ##10s

# for i, chunk in enumerate(chunks):
#     ##ichunk
#     chunk_name = "section{0}.acc".format(i)
#     print(chunk_name)
#     ##
#     chunk.export(chunk_name, format="acc")
```
